```python
# Q4
color = input("Enter color(red,yellow,green)")
if color == "":
    print("cannot be empty")
else:
    match color:
        case "red":
            print("stop")
        case "yellow":
            print("get ready")
        case "green":
            print("go")
        case _:
            print("unkown color")
```

```python
# Q5
print("1-Addition")
print("2-Subtraction")
print("3-Multiplication")
print("4-Division")

choise = input("choose (1-4) ")
num1 = float(input("Enter first number"))
num2 = float(input("Enter second number"))

match choise:
    case "1":
        print("Result:", num1 + num2 )
    case "2":
        print("Result:", num1 - num2 )
    case "3":
        print("Result:", num1 * num2 )
    case "4":
        print("Result:", num1 / num2 )
        
```

```python
score = int(input("Enter student score"))
if score >= 50:
    print("passed")
else:
    print("failed")

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 50:
    grade = "D"
else:
    grade = "F"

match grade:
    case "A":
        print("Excellent")
    case "B":
        print("very Good")
    case "C":
        print("Good")
    case "D":
        print("pass")
    case "F":
        print("Fail")
```
